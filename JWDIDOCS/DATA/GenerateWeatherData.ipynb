```python
conditions = ["Raining", "Fire", "Cyclone", "Sunny", "Clowdy", "Partly Clowdy", "Stormy"]
```

```python
moon_phase = ["Waning Crescent", "Full", "No Moon", "Partial Moon"]
```

```python
import random

generated_df = []
for x in range(0, 1000):
    new_list = []
    new_list.append(dt.datetime.now() + dt.timedelta(days=x))
    new_list.append(1740614400)
    new_list.append(8)
    new_list.extend( [random.random()*100 for x in range(10)])
    new_list.extend( [random.randint(0, 50) for x in range(8)])
    new_list.append( conditions[random.randint(0,6)])
    new_list.append(f'//cdn.weatherapi.com/weather/64x64/day/{str(random.randint(100, 200))}.png')
    new_list.extend( [random.randint(0, 50) for x in range(6)])
    new_list.append(moon_phase[random.randint(0,2)])
    new_list.append(random.randint(0,20))
    generated_df.append(new_list)
    
df=pd.DataFrame(generated_df,columns=input_example.columns)
random.random()*100
```
